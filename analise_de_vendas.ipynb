```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo o dataframe e exibindo as 5 primeiras linhas
df = pd.read_csv("Online Sales Data.csv")
df.head()
```

```python
# Descrição estatística do DataFrame
df.describe()
```

```python
# Verificando valores nulos
df.isnull().sum()  # Sem valores nulos
```

```python
# Verificando valores duplicados
df.duplicated().sum()  # Sem dados duplicados
```

```python
# Produto mais vendido
produto_mais_vendido = df.groupby('Product Name')['Units Sold'].sum().idxmax()  # Nome do produto mais vendido
quantidade_produto_mais_vendido = df.groupby('Product Name')['Units Sold'].sum().max()  # Quantidade de unidades vendidas do produto mais vendido

print(f"O produto que mais vendeu foi {produto_mais_vendido}")
print(f"Foram vendidas {quantidade_produto_mais_vendido} unidades")
```

```python
# Categoria de produto mais vendido
categoria_mais_vendida = df.groupby('Product Category')['Units Sold'].sum()
df_categoria_mais_vendida = pd.DataFrame(categoria_mais_vendida)

# Plotando o gráfico de barras das categorias de produtos mais vendidas
plt.figure(figsize=(16,6))
df_categoria_mais_vendida.plot(kind='bar', color='lightgreen')
plt.xticks(rotation=45)
plt.title('Categorias de Produto Mais Vendidas')  # Título do gráfico
plt.xlabel('Categoria de Produto')  # Rótulo do eixo x
plt.ylabel('Unidades Vendidas')  # Rótulo do eixo y
plt.show()
```

```python
# Histograma de Unidades Vendidas
unidades_vendidas = df[['Units Sold']]
plt.figure(figsize=(12,6))
plt.hist(unidades_vendidas, bins=20)
plt.xlabel('Unidades Vendidas')  # Rótulo do eixo x
plt.ylabel('Frequência')  # Rótulo do eixo y
plt.title('Distribuição das Unidades Vendidas')  # Título do histograma
plt.grid(True)
plt.show()
```

```python
# Histograma do Preço Unitário
preco_unitario = df[['Unit Price']]
plt.figure(figsize=(12,6))
plt.hist(preco_unitario, bins=20)
plt.xlabel('Preço Unitário')  # Rótulo do eixo x
plt.ylabel('Frequência')  # Rótulo do eixo y
plt.title('Distribuição do Preço Unitário')  # Título do histograma
plt.show()
```

```python
# Histograma da Receita Total
receita_total = df[['Total Revenue']]
plt.figure(figsize=(12,6))
plt.hist(receita_total, bins=20)
plt.xlabel('Receita Total')  # Rótulo do eixo x
plt.ylabel('Frequência')  # Rótulo do eixo y
plt.title('Distribuição da Receita Total')  # Título do histograma
plt.show()
```

```python
# Região com mais vendas por unidade
vendas_por_regiao = df.groupby('Region')['Units Sold'].sum()
df_vendas_por_regiao = pd.DataFrame(vendas_por_regiao)
plt.figure(figsize=(8,8))
plt.pie(df_vendas_por_regiao['Units Sold'], labels=df_vendas_por_regiao.index, autopct='%1.1f%%')
plt.title('Vendas por Região')  # Título do gráfico de pizza
plt.show()
```

```python
# Região que mais gerou receita
receita_por_regiao = df.groupby('Region')['Total Revenue'].sum()
df_receita_por_regiao = pd.DataFrame(receita_por_regiao)
plt.figure(figsize=(8,8))
plt.pie(df_receita_por_regiao['Total Revenue'], labels=df_receita_por_regiao.index, autopct='%1.1f%%')
plt.title('Receita por Região')  # Título do gráfico de pizza
plt.show()
```
